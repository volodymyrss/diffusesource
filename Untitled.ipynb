```python
import numpy
import astropy.units as u
from astropy.io import fits
import csv
from scipy.interpolate import interp1d
from astropy.coordinates import SkyCoord
from gammapy.irf import load_cta_irfs
from gammapy.maps import WcsGeom, MapAxis
from gammapy.modeling.models import (
    PowerLawSpectralModel,
    GaussianSpatialModel,
    SkyModel,
    Models,
    FoVBackgroundModel,
    TemplateSpatialModel,
    ConstantSpectralModel,
    PowerLawNormSpectralModel,
)
from gammapy.maps import MapAxis, WcsGeom
from gammapy.makers import MapDatasetMaker, SafeMaskMaker
from gammapy.modeling import Fit
from gammapy.data import Observation
from gammapy.datasets import MapDataset
from gammapy.maps import Map
import matplotlib.pyplot as plt
```

```python
# Loading IRFs
irfs = load_cta_irfs("/Users/sokolenko/Dropbox/irf_file.fits")

# Define the observation parameters (typically the observation duration and the pointing position):
livetime = 10.0 * u.hr
pointing = SkyCoord(98.48, 17.77, unit="deg", frame="icrs")

energy_axis = MapAxis.from_edges(
    numpy.logspace(1.4771212547, 5, 51)/1000, unit="TeV", name="energy", interp="log"
)

geom = WcsGeom.create(
    skydir=(98.48, 17.77),
    binsz=0.05,
    width=(10, 10),
    frame="icrs",
    axes=[energy_axis],
)

empty = MapDataset.create(geom, name="dataset-simu")
```

```python
filename = "allenergies_TeV.fits"
m = Map.read(filename)
# m = m.copy(unit="1 / sr")

# spectral_model = ConstantSpectralModel(const="1 / (cm2 s TeV)",normalize=False)
# spatial_model = TemplateSpatialModel(m, filename=filename, normalize=False)

# model = SkyModel(spectral_model=spectral_model, spatial_model=spatial_model, name="pwl-template-model")

# m.plot_grid()

m = m.copy(unit="1 / (cm2 TeV s sr)")

spatial_model = TemplateSpatialModel(m, filename=filename, normalize=False)

Geminga_model = SkyModel(spectral_model=PowerLawNormSpectralModel(), spatial_model=spatial_model, name="Geminga-template-model")

m.plot_grid()
```

```python
# bkg_spectral_model = PowerLawSpectralModel(
#     index=2.7, amplitude="1e6 cm-2 s-1 TeV-1", reference="1 TeV")

bkg_model = FoVBackgroundModel(dataset_name="dataset-simu")
#bkg_model = FoVBackgroundModel(spectral_model=bkg_spectral_model, dataset_name="dataset-simu")

#models = Models([bkg_model])
models = Models([Geminga_model,bkg_model])
print(models)

print(models.to_yaml())
```

```python
obs = Observation.create(pointing=pointing, livetime=livetime, irfs=irfs)
print(obs)
```

```python
# Open the FITS file
hdul = fits.open('observation_2d.fits')

# Print the information about the file
hdul.info()

counts = hdul[1].data
background = hdul[5].data
bands = hdul[2].data
#npbands = numpy.array(bands).reshape(50,1)
counts_2d = counts.reshape(50, -1)
background_2d = background.reshape(50,-1)
counts_sum = []
background_sum = []
for i in range(0,50):
    counts_sum.append(sum(counts_2d[i]))
    background_sum.append(sum(background_2d[i]))
```

```python
emin_cta, flux_cta = numpy.loadtxt('CTA_point_source_left.txt',unpack=True) # E[TeV], E^2 * Flux Sensitivity [erg cm^2/s]
emax_cta, flux_cta = numpy.loadtxt('CTA_point_source_right.txt',unpack=True)
emean_cta, flux_cta = numpy.loadtxt('CTA_point_source_mean.txt',unpack=True)

energy_Flux, Flux  = numpy.loadtxt('alpha15_model_Flux.dat',unpack=True)  #E[GeV], Flux [eV/cm^2/s]

energy_Flux=energy_Flux/10**3 # E[TeV] 

x_HAWC_u = []
y_HAWC_u = []
with open('HAWC_upper.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x_HAWC_u.append(float(row[0]))
        y_HAWC_u.append(float(row[1])) 

x_HAWC_l = []
y_HAWC_l = []
with open('HAWC_lower.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x_HAWC_l.append(float(row[0]))
        y_HAWC_l.append(float(row[1]))
        
x_HESS_u = []
y_HESS_u = []
with open('HESS_upper.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        x_HESS_u.append(float(row[0]))
        y_HESS_u.append(float(row[1])) 

x_HESS_l = []
y_HESS_l = []
with open('HESS_lower.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=' ')
    for row in plots:
        x_HESS_l.append(float(row[0]))
        y_HESS_l.append(float(row[1]))

x_Magic = []
y_Magic = []
with open('Magic.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x_Magic.append(float(row[0]))
        y_Magic.append(float(row[1]))
        
x_HAWC_u = numpy.array(x_HAWC_u)/10**3
y_HAWC_u = numpy.array(y_HAWC_u) * 1.602*10**-12
x_HAWC_l = numpy.array(x_HAWC_l)/10**3
y_HAWC_l = numpy.array(y_HAWC_l) * 1.602*10**-12
x_Magic = numpy.array(x_Magic)/10**3
y_Magic = numpy.array(y_Magic) * 1.602*10**-12
                
Int_HAWC_u=interp1d(x_HAWC_u, y_HAWC_u,fill_value="extrapolate")
Int_HAWC_l=interp1d(x_HAWC_l, y_HAWC_l,fill_value="extrapolate")

Int_HESS_u=interp1d(x_HESS_u, y_HESS_u,fill_value="extrapolate")
Int_HESS_l=interp1d(x_HESS_l, y_HESS_l,fill_value="extrapolate")
```

```python
def in_in_out(left,right,quantity): #selects energies of the old bins that are between left and right energy value
#res=[quantity[i] for i in range(len(quantity)) if quantity[i]>=right and quantity[i]<left]
    res=[i for i in range(len(quantity)) if quantity[i]>=right and quantity[i]<left]
    return(res)

def bin_group(input_bins, desired_bins): #selects indecies of the old binning that lie inside the new bins
    input_bins_in_desired_bins=[in_in_out(desired_bins[i+1],desired_bins[i], input_bins) for i in range(len(desired_bins)-1)]
    return input_bins_in_desired_bins

def bin_restructurize(input_bins, input_quantity, desired_bins): #rebinning input quantity to the new bins
    indexes=bin_group(input_bins, desired_bins)
    output_quantity=[]
    for i in range(len(desired_bins)-1):
        sum_bin=0
        if len(indexes[i])>0:
            for j in range(len(indexes[i])):
                sum_bin+=input_quantity[indexes[i][j]]
            output_quantity+=[sum_bin/len(indexes[i])]   
    return output_quantity

def bin_restructurize(input_bins, input_quantity, desired_bins): #rebinning input quantity to the new bins
    indexes=bin_group(input_bins, desired_bins)
    output_quantity=[]
    for i in range(len(desired_bins)-1):
        sum_bin=0
        if len(indexes[i])>0:
            for j in range(len(indexes[i])):
                sum_bin+=input_quantity[indexes[i][j]]
            output_quantity+=[sum_bin/len(indexes[i])]   
    return output_quantity
```

```python
#rebinning

CR_N5_rebinned=bin_restructurize(energy_Flux,numpy.array(background_sum),emin_cta)[2:]
source_N5_rebinned=bin_restructurize(energy_Flux,numpy.array(counts_sum),emin_cta)[2:]
Flux_rebinned=bin_restructurize(energy_Flux,Flux,emin_cta)[2:]

len(Flux_rebinned),len(CR_N5_rebinned),len(source_N5_rebinned),len(emin_cta[2:-1])
```

```python
# Sensitivity (5 sigma)

StoN = numpy.array(source_N5_rebinned) / (numpy.array(CR_N5_rebinned))**0.5

Sensitivity = numpy.array(Flux_rebinned)*u.eV/u.cm**2/u.s * 5/StoN # [eV/cm^2/s]

Sensitivity=Sensitivity.to(u.erg/u.cm**2/u.s)
```

```python
plt.errorbar(emean_cta[2:-1], Sensitivity,xerr=(emax_cta[2:-1]-emean_cta[2:-1])/2,label='alpha15 10h')
plt.errorbar(emean_cta[2:-1],flux_cta[2:-1],label='CTA point source 50h')
plt.fill_between(x_HAWC_l, Int_HAWC_u(x_HAWC_l)/7.6,Int_HAWC_l(x_HAWC_l)/7.6, color='green',alpha=.4,label='HAWC (scaled)')
plt.fill_between(x_HESS_l, Int_HESS_u(x_HESS_l),Int_HESS_l(x_HESS_l), color='red',alpha=.4,label='HESS')
plt.scatter(x_Magic,y_Magic,color = 'purple',label='Magic')
plt.loglog()
plt.xlabel("$E, [TeV]$")
plt.ylabel("Sensitivity, [erg cm$^{-2} \cdot s^{-1}$]")
plt.legend(frameon=False, loc='upper right')
```

```python
# Estimation
e_eff_area_cta, eff_area_cta = numpy.loadtxt('effective_area_CTA_S_50h.dat',unpack=True) # E[TeV], Eff. area [m^2]
source_N5_estimation_tmp = Flux / (energy_Flux*10**12)  * (10*60*60) * numpy.interp(energy_Flux,e_eff_area_cta,eff_area_cta) * 100**2
source_N5_estimation = numpy.interp(energy_Flux,energy_Flux,source_N5_estimation_tmp)

source_N5_rebinned_estimation=bin_restructurize(energy_Flux,source_N5_estimation,emin_cta)[2:]

StoN_estimation = numpy.array(source_N5_rebinned_estimation) / (numpy.array(CR_N5_rebinned))**0.5

Sensitivity_estimation = numpy.array(Flux_rebinned)*u.eV/u.cm**2/u.s * 5/StoN_estimation # [erg/cm^2/s]

Sensitivity_estimation=Sensitivity_estimation.to(u.erg/u.cm**2/u.s)
```

```python
plt.errorbar(emean_cta[2:-1], Sensitivity,xerr=(emax_cta[2:-1]-emean_cta[2:-1])/2,label='alpha15 10h')
plt.errorbar(emean_cta[2:-1], Sensitivity_estimation,xerr=(emax_cta[2:-1]-emean_cta[2:-1])/2,label='estimation')
plt.errorbar(emean_cta[2:-1],flux_cta[2:-1],label='CTA point source 50h')
plt.fill_between(x_HAWC_l, Int_HAWC_u(x_HAWC_l)/7.6,Int_HAWC_l(x_HAWC_l)/7.6, color='green',alpha=.4,label='HAWC (scaled)')
plt.fill_between(x_HESS_l, Int_HESS_u(x_HESS_l),Int_HESS_l(x_HESS_l), color='red',alpha=.4,label='HESS')
plt.scatter(x_Magic,y_Magic,color = 'purple',label='Magic')
plt.loglog()
plt.xlabel("$E, [TeV]$")
plt.ylabel("Sensitivity, [erg cm$^{-2} \cdot s^{-1}$]")
plt.legend(frameon=False, loc='upper right')
```
